```python
import pandas as pd 
for i in range(len(data_list)):
    mat = scipy.io.loadmat(data_list[i]) # load ith mat file 
    for j in range(10):              # load jth block 
        data = mat['run'][0,j][0][0][0]     # ith file and jth block's signals
        #data = a.transpose           
        df = pd.DataFrame(data = data, columns = columns) # make a data frame
        df.to_csv('BNCI_ERPs%s%s.csv' % (i,j))                     # save it to csv file
```
